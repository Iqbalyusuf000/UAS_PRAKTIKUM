```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from google.colab import files
uploaded = files.upload()
```

```python
dataset = pd.read_csv('stock1.csv')
dataset.head(20)
```

**(A1) Distribusi Data Atribut Purchase Price**

```python
#Membuat ukuran figure (panjang, tinggi)
plt.figure(figsize = (10,5))

#Membuat Histogram Chart
plt.hist(dataset['Purchase Price ($)'],bins = 15, edgecolor = 'white', color = 'darkblue', histtype = 'barstacked',
         align = 'right', label = "Purchase Price")

#Membuat Judul dan Label
plt.title('Distribusi Data Pada Atribut Purchase Price', color='red', fontsize=15, fontname = 'Monospace')
plt.xlabel('Purchase Price (Harga Beli)', color='red', fontsize=11)
plt.ylabel('Frekuensi', color='red',fontsize=11)

#Menampilkan Perintah
plt.show()
```

**(A2) Distribusi Data Atribut Current Price**

```python
#Membuat ukuran figure (panjang, tinggi)
plt.figure(figsize = (10,5))

#Membuat Histogram Chart
plt.hist(dataset['Current Price ($)'],bins = 15, edgecolor = 'white', color = 'orange', histtype = 'barstacked',
         align = 'right', label = "Current Price")

#Membuat Judul dan Label
plt.title('Distribusi Data Pada Atribut Current Price', color='magenta', fontsize=15, fontname = 'Monospace')
plt.xlabel('Current Price (Harga Terkini)', color='magenta', fontsize=11)
plt.ylabel('Frekuensi', color='magenta',fontsize=11)

#Menampilkan Perintah
plt.show()
```

**(A3) Distribusi Data Atribut Gain Loss**

```python
#Membuat ukuran figure (panjang, tinggi)
plt.figure(figsize = (10,5))

#Membuat Histogram Chart
plt.hist(dataset['Gain / Loss ($)'],bins = 15, edgecolor = 'white', color = 'seagreen', histtype = 'barstacked',
         align = 'right', label = "Gain / Loss")

#Membuat Judul dan Label
plt.title('Distribusi Data Pada Atribut Gain / Loss', color='tomato', fontsize=15, fontname = 'Monospace')
plt.xlabel('Gain / Loss (Rugi)', color='tomato', fontsize=11)
plt.ylabel('Frekuensi', color='tomato',fontsize=11)

#Menampilkan Perintah
plt.show()
```

**(B) Atribut Stock dan Symbol dengan Dividend / Share Top 5**

```python
# Membuat Urutan dari Terbesar ke Kecil
urutan_dataset = dataset.sort_values('Dividend / Share ($)', ascending = False)

# Membuat Data Top 5
top5_dividend = urutan_dataset.head(5)

# Memberikan Nomor urut
top5_dividend = top5_dividend.set_index(pd.Index(range(1, len(top5_dividend)+1)))

#Memilih Kolom Yang di Tampilkan
top5 = top5_dividend[['Stock Name', 'Symbol', 'Dividend / Share ($)']]

#Menampilkan Perintah
print('Top 5 Dividend / Share (dalam $) Beratribut Stock Name & Symbol')
print(top5)
```

**(C) Atribut Stock dan Symbol dengan Market Value Top 5**

```python
# Membuat Urutan dari Terbesar ke Kecil
urutan_dataset = dataset.sort_values('Market Value ($)', ascending = False)

# Membuat Data Top 5
top5_market = urutan_dataset.head(5)

# Memberikan Nomor urut
top5_market = top5_market.set_index(pd.Index(range(1, len(top5_dividend)+1)))

#Memilih Kolom Yang di Tampilkan
top5 = top5_market[['Stock Name', 'Symbol', 'Market Value ($)']]

#Menampilkan Perintah
print('Top 5 Market Value (dalam $) Beratribut Stock Name & Symbol')
print(top5)
```
